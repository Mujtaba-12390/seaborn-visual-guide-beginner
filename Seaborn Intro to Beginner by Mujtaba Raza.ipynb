# Relational Plots and Subplots

Many questions in data science are centered around describing the relationship between two quantitative variables.

Seaborn calls plots that visualize this relationship "relational plots"

# For Example :

# Introducing relplot() 

"relplot" stands for "relational plot" it enables you to visualize relationship between two quantitative variables using either scatter plot or line plot



# scatterplot() Vs relplot()

```python
tips=pd.read_csv("tips.csv")
tips.head(10)
```

## Using scatterplot()

```python
sns.scatterplot(x = "total_bill",\
                y = "tip", \
                data=tips)

plt.show()
```

## Using relplot()

```python
sns.relplot(x = "total_bill",\
                y = "tip", \
                kind="scatter",\
                data=tips)

plt.show() 
#If we see the replot and scatter plot there is no such similarity except removing of right spine.
# According to me replot shows good and nice comparision and its help us to create multiples sub plot
```

# Subplotting

## row

```python
sns.relplot(x = "total_bill",\
                y = "tip", \
                kind="scatter",\
                row="sex",\
                data=tips) #after analysis different factors I visualize that tip has correlation with day,time, sex 
                            # as compare to size and smoker

plt.show() 
```

## column

```python
sns.relplot(x = "total_bill",\
                y = "tip", \
                kind="scatter",\
                col="size",\
                data=tips,\
                col_wrap=2,\
                hue="sex")

plt.show()
```

# Challenge :


    1) Make a scatter plot of the day of the week column (investigate which day has the highest amount of tips)
    
    2) Affect Of Gender On Tip & Bill Amount

```python
sns.set_style("dark")
sns.color_palette("bright")
sns.set_context("notebook")

g=sns.relplot(x = "total_bill",\
                y = "tip", \
                kind="scatter",\
                col="day",\
                data=tips,\
                col_wrap=2,\
                hue="sex")

g.fig.set_figheight(8)
g.fig.set_figwidth(15)

plt.show() 
```

# Observation: 

    Day of the week with high amount of tips : Saturday

    Gender that spends more money in the restaurant and give more tips : Male
